```python
print('開始心理測驗吧～～')
print('五題測出生活重心該放哪裡，讓你有滿滿正能量，一路過關斬將、化危機為轉機！')
start=input('輸入1開始，任意鍵為結束：')
while start == '1':
    
    while True:
        print()
        print('1：以下什麼對你來說最重要?')
        one =input(' a.事業\n b.愛情\n c.友情\n d.家庭\n 請選擇：')
        if (one =='a') or (one =='b') or (one =='c') or (one =='d'):
            break
        else:
            print("sorry無此選項，請重新輸入")
            
    
    while True:
        print()
        print('2：在路上遇到人搭訕，你覺得他是...?')
        two = input(' a.想要認識你，留聯絡方式\n b.來向你推銷產品\n c.迷路了，只是想問路\n d.遇到困難，需要幫助\n 請選擇：')
        if (two =='a') or (two =='b') or (two =='c') or (two =='d'):
            break
        else:
            print("sorry無此選項，請重新輸入")

    while True:
        print()
        print('3：覺得自己像哪種動物?')
        three = input(' a.獅子\n b.貓頭鷹 \n c.鳥\n d.狗\n 請選擇：')
        if (three =='a') or (three =='b') or (three =='c') or (three =='d'):
            break
        else:
            print("sorry無此選項，請重新輸入")

    while True:
        print()
        print('4：夢裡在BUFFET用餐，你拿了什麼菜?')
        four = input(' a.整盤都是肉類和海鮮\n b.健康的青菜水果類\n c.鹹、甜、冷、熱都拿一點\n d.各式各樣的甜點\n 請選擇：')
        if (four =='a') or (four =='b') or (four =='c') or (four =='d'):
            break
        else:
            print("sorry無此選項，請重新輸入")

    while True:
        print()
        print('5：路邊有隻貓咪在對你撒嬌，你會？')
        five = input(' a.遠遠看他幾眼就離去\n b.買罐頭來給他吃\n c.摸摸他、陪他玩一下\n d.幫他拍張照片上傳IG\n 請選擇：')
        if (five =='a') or (five =='b') or (five =='c') or (five =='d'):
            break
        else:
            print("sorry無此選項，請重新輸入")



    dict={'a':'5','b':'4','c':'3','d':'2'}
    one=dict.get(one)
    two=dict.get(two)
    three=dict.get(three)
    four=dict.get(four)
    five=dict.get(five)
    score=int(one)+int(two)+int(three)+int(four)+int(five)

    print()
    print('您的生活重心該放在：')
    if 23<=int(score)<=25:
        print('事業')
        print('是適合在工作上專心衝刺的一年，可能有難得的升遷或是工作機會。應主動積極爭取表現機會。另外，平時要多努力人脈累積，避免和同事在合作中起衝突。如果正在猶豫是否要換工作，可以再緩緩等待時機，會有貴人提拔相助。')
    elif 20<=int(score)<=22:
        print('愛情')
        print('沒有對象的你，將在今年展開幾段情感關係，有機會遇到能長遠發展的好對象。不過，也要注意需果斷斬掉爛桃花，避免浪費時間和金錢，對精神消耗太大！已有穩定對象的你，感情會面臨大幅度的改變，也許是加溫昇華進入下一個階段，也有可能會面臨重大爭吵，需多留意日常相處細節，讓對方感受到你的用心。')
    elif 17<=int(score)<=19:
        print('生活目標')
        print('感覺蠟燭兩頭燒，忙碌到覺得有些迷惘的一年。靜下心重新檢視你真正渴望的事情，設定好主要目標後，就屏除雜念專心往目標前進，果斷放下其他瑣事吧！也可以好好安排趟旅遊充電一下，同時幫助自己找回目標方向。')
    elif 14<=int(score)<=16:
        print('課業、進修')
        print('趁著沒有太多起伏、相對平穩的今年，努力充實自我、培養實力，保持積極學習的態度。還在念書的你，可以試著爭取國際交換學生的機會，增強自己的語文能力。已經在工作的上班族，則應透過妥善的時間安排，好好進修、培養第二專長。')
    else:
        print('家庭')
        print('和家庭相關的事務增多，要付出更多心力在家庭關係經營。可能會和家人在相處上發生不愉快的情形，讓你感到相當心煩。同時，家人的身體狀況也需要你花更多心思照顧。嘗試多花點時間耐心溝通，衝突將會迎刃而解。')
    print()
    start=input('想再測一次嗎～輸入1開始，任意鍵為結束：')

```
