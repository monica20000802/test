## 莎士比亞詩集翻譯

```python
#開啟'poem.txt'檔，把原文詩寫進檔案中並關閉
with open('poem.txt','w+',encoding='utf-8') as file:
    file.write('<Sonnet XVIII Shall I Compare Thee To a Summer’s Day>\n')
    file.write('Shall I compare thee to a Summers day?\n')
    file.write('Thou art more louely and more temperate:\n')
    file.write('Rough windes do fhake the darling buds of Maie,\n')
    file.write('And Sommers leafe hath all too fhort a date:\n')
    file.write('Sometime too hot the eye of heauen fhines,\n')
    file.write('And often is his gold complexion dimm’d,\n')
    file.write('And cuery faire from faire fome-time declines,\n')
    file.write('By chance,or natures changing courfe rntrimd:\n')
    file.write('But thy eternall Sommer fhall not fade,\n')
    file.write('Nor loofe poffeffion of that faire thou ow’ft,\n')
    file.write('Nor fhall death brag thou wandr’ft in his fhade,\n')
    file.write('When in eternall lincs to time thou grow’ft,\n')
    file.write('So long as men can breath or eyes can fee,\n')
    file.write('So long liues this,and this giues life to thee\n')
    
#開啟一個檔案，把檔案中的資料都讀進一個串列中後關閉
file1=open('poem.txt', 'r', encoding='utf-8')
list_lines=file1.readlines()
file1.close()

#串列內的元素一行一行印出來，印出完整的原文詩
for line in list_lines:
    print(line, end='')

#請使用者輸入原文中欲查詢之單字，並印出翻譯後的現在的用字（原文）
dic={'Thee':'you;你（受格）','thou':'you;你（主格）','louely':'lovely; 可人的',
     'more lovely':'lovelier; 可人的（比較級）','windes':'winds; 風','fhake':'shake;搖晃',
     'Maie':'may;五月','somme':'summer;夏天','leafe':'lease;租約','hath':'has',
     'fhort':'short;短暫的','heauen':'heaven;天堂','fhines':'shines;閃耀',
     'dimm’d':'dimmed;暗淡的','euery':'every;每','faire':'beautiful;美麗的',
     'courfe':'course;行徑','vntrim’d':'untrimmed;未經修飾的','thy':'your; 你的（屬格）',
     'eternall':'eternal;永遠的','fhall':'shall;應該','loofe':'lose;失去',
     'poffeffion':'possession;擁有','ow’ft':'owned;擁有的','wand’rest':'wandering;漫遊於',
     'fhade':'shade;陰影','grow’ft':'grow;誕生','fee: see':'see;看到',
     'liues':'lives;長存','giues':'gives;給予'}
print()
while True:  
    find=input('請輸入要查詢的單字，結束查詢請輸入"done":')
    if find!= "done":
        print(dic.get(find,'查詢不到此單字'))
    else:
        print()
        print('英/中文的翻譯版:') #印出英/中文的翻譯版
        print('<Sonnet XVIII Shall I Compare Thee To a Summer’s Day>\n',
              'Shall I compare you to a summer’s day?\n',
              '我是否該將你比喻為夏日？\n',
              'You are lovelier and milder.\n',
              '比起夏日，你更溫柔動人。\n',
              'Rough winds shake the pretty buds of May.\n',
              '狂風搖落五月珍愛的花蕊，\n',
              'And summer doesn’t last nearly long enough.\n', 
              '夏日只是個稍縱即逝的季節。\n',
              'Sometime the sun shines too hot\n',
              '時而，穹蒼之眼炙熱難當，\n',
              'And often its golden face is darkened by clouds.\n',
              '時常，穹蒼容顏金光褪藏，\n',
              'And everything beautiful stops being beautiful,\n',
              '美景時時衰微，融入下一個美景，\n',
              'Either by accident or simply in the course of nature.\n',
              '或順應機緣，或循自然荒野的行徑。\n',
              'But your eternal summer will never fade,\n',
              '然而，那屬於你的永恆夏日從不老去，\n',
              'Nor will you lose your possession of beauty\n',
              '你所擁有的夏日之美也不曾淡去，\n',
              'Nor shall Death brag that you are wandering in the underworld.\n',
              '死亡無法誇言，你曾漫遊於他的陰影，\n',
              'Once you’re captured in my eternal verses,\n',
              '因為你已誕生於時間永恆的詩句。\n',
              'As long as men are alive and have eyes with which to see,\n',
              '只要有人呼吸，有眼凝視著，\n',
              'This poem will live and keep you alive.\n',
              '這首詩將長存，並賜予你永生。')
        break
```
