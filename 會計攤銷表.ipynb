## 會計攤銷表

```python
#input 發行價格、債券面值、年期、每年付息次數、票面利率、有效利率，並檢查格式與疏漏
while True:
    issued_price=input('請輸入發行價格:')
    if (issued_price.isdigit()==False) or (float(issued_price)<0):
        print("格式錯誤，請重新輸入")
    else:
        issued_price=int(issued_price)
        break
while True:
    face_value=input('請輸入債券面值:')
    if (face_value.isdigit()==False) or (float(face_value)<0):
        print("格式錯誤，請重新輸入")
    else:
        face_value=int(face_value)
        break
                
while True:
    years=input('請輸入債券年期:')
    if (years.isdigit()==False) or (float(years)<0) :
        print("格式錯誤，請重新輸入")
    else:
        years=int(years)
        break
while True:
    frequency=input('請輸入每年付息次數:')
    if (frequency.isdigit()==False) or (float(frequency)<0) :
        print("格式錯誤，請重新輸入")
    else:
        frequency=int(frequency)
        break
        
while True:
    par_i=input('請輸入票面利率(%):')
    if (par_i.isdigit()==False) or (float(par_i)<0):
        print("格式錯誤，請重新輸入")
    else:
        break
 
while True:
    market_i=input('請輸入有效利率(%):')
    if (market_i.isdigit()==False) or (float(market_i)<0):
        print("格式錯誤，請重新輸入")
    if(int(issued_price)>int(face_value)) and (float(par_i)<float(market_i)):
        print("票面利率要大於有效利率，請重新輸入")#溢價發行，票面利率一定要>有效利率
    elif(int(issued_price)<int(face_value)) and (float(par_i)>float(market_i)):
        print("票面利率要小於有效利率，請重新輸入")#折價發行，票面利率一定要<有效利率
    else:
        market_i=int(market_i)/100
        par_i=int(par_i)/100
        break
    
#宣告變數
#計算期數、實付利息
i_exp=利息費用=0
amortize=每年攤銷費用=0
times=期數=years*frequency
cash_paid=實付利息=round(face_value*par_i/frequency)
carry_amount=issued_price

#開啟名為攤銷表的檔案，用write()&格式化字串寫出表頭及第一列入檔案
file1=open('攤銷表.txt', 'w+', encoding='utf-8')
file1.write('%-5s%5s%10s%10s'%('期數','利息費用','溢價攤銷額','帳面金額\n'))
file1.write(f"{0:4d}{carry_amount:40d}\n")

#將期數當成計數器，計算期數、利息費用、溢價攤銷額、帳面金額，一期一行用write()&格式化字串寫入檔案
#輸出的數字要四捨五入到整數
for amo in range(times-1):
    i_exp=round(carry_amount*market_i)
    amortize=round(cash_paid-i_exp)
    carry_amount-=round(amortize)
    file1.write(f"{amo+1:4d}{i_exp:12d}{amortize:15d}{carry_amount:13d}\n")
    
#最後一期需調整數字，寫入檔案後並關閉
amortize=round(carry_amount-face_value)
i_exp=round(cash_paid-amortize)
carry_amount=face_value
file1.write(f"{5:4d}{i_exp:12d}{amortize:15d}{carry_amount:13d}\n")
file1.close()

print()
#開啟一個檔案，把檔案中的資料都讀進一個串列中後關閉
file2=open('攤銷表.txt', 'r', encoding='utf-8')
list_lines=file2.readlines()
file2.close()

#串列內的元素一行一行印出來
for line in list_lines:
    print(line, end='')

```
